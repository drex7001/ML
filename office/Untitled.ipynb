```python
import pandas as pd

# Load the data from the CSV files
zig_old = pd.read_csv('zig_old.csv')

# Display the first few rows of zig_old
zig_old.head()
```

```python
# Filter rows based on the given location criteria
filtered_zig_old = zig_old[zig_old['Location'] == "Online Operation - Charles Drive"]

# Drop unnecessary columns
filtered_zig_old = filtered_zig_old.drop(columns=[col for col in filtered_zig_old.columns if 'Unnamed' in col])

filtered_zig_old.head()
```

```python
# Load the data from the zig_new.csv file
zig_new = pd.read_csv('zig_new.csv')

# Display the first few rows of zig_new
zig_new.head()
```

```python
# Update stock counts for each SKU in zig_old in zig_new
for index, row in filtered_zig_old.iterrows():
    sku = row['SKU']
    mask = zig_new['SKU'] == sku
    zig_new.loc[mask, ['Incoming', 'Unavailable', 'Committed', 'Available', 'On hand']] = row[['Incoming', 'Unavailable', 'Committed', 'Available', 'On hand']].values

# Display the first few rows of zig_new after updating
zig_new.head()
```
